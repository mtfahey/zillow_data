```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("../data/source/Zillow Data for Scripps_Landlords Listing for Sale.xlsx")
data
```

### Make all charts

```python
for metric in data["metric"].unique(): 
    this_data = data[data["metric"]==metric]
    for use in this_data["usecode"].unique():
        this_data_use = this_data[this_data["usecode"]==use]
        name = metric+"_"+use
        this_data_use.plot(x="date", y="val_final", title=name)
```
